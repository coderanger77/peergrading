## Libraries and functions 


```python
import pandas as pd 
import numpy as np
from matplotlib import pyplot as plt

def limit(x,min,max):
    if(x<min):
        return min
    if(x>max):
        return max
    return x

```

## Initializations 


```python
df_true=pd.DataFrame(columns=['s','b','t'])
student_count=100
minmarks=0
maxmarks=100
no_of_grades=10
mu_true=60
sigma_true=15
beta_true=10  #beta is actually 10. so when we use 1/beta , we are using 0.1 instead

lambda_prior=1



## intial priors for all the students.[beta,eta,mu,gama]

initial_beta=1
initial_eta=1
initial_mu= 50
initial_gama = 0.03333333333
```

## ** True data (Simulator) **

```python
#generate true values
df_true=pd.DataFrame(columns=['id','s','b','t'])

for u in range(0,student_count):   
    #s_u true scores for each grader 
    su=np.random.normal(mu_true,sigma_true,1)[0]
    su=limit(su,minmarks,maxmarks)
        
    #bias of each user ranges from 0-
    bu=np.random.normal(0,no_of_grades,1)[0]
    bu=limit(bu,-no_of_grades,no_of_grades)
    
    #t -> reliability
    tu=np.random.normal(su,beta_true,1)[0]
    
    df_true.loc[u]=[int(u),su,bu,tu]
    
df_true['id']=df_true['id'].astype('int')

print(df_true.describe())
#min value of s is only 24 
#df_true.dtypes
#df_true
```

## Allocation algorithm
